Exercise 2

(a).

```python
lambda_1 = [0, 10 ** -6, 10 ** -5, 10 ** -4, 10 ** -3, 10 ** -2, 10 ** -1, 1]
lambda_2 = [0, 10 ** -6, 10 ** -5, 10 ** -4, 10 ** -3, 10 ** -2, 10 ** -1, 1]
param = pd.DataFrame(columns=['lambda_1','lambda_2','alpha','l1_ratio'])
for i in lambda_1:
  for d in lambda_2:
    alpha, l1_ratio = cal_lambdas(i,d)
    temp = {'lambda_1': i, 'lambda_2': d, 'alpha': alpha, 'l1_ratio': l1_ratio} 
    param = param.append(temp, ignore_index=True)

def cal_lambdas(L1,L2):
  alpha = L1+L2
  if L1 == 0 and L2 == 0:
    l1_ratio = 0
  else:
    l1_ratio = L1/(L1+L2)
  return(alpha, l1_ratio)

#param
```

(b,c,d).

```python
Train_X = Train.iloc[:, 0:81]
Train_Y = Train.iloc[:, 81]
Validation_X = Validation.iloc[:,0:81]
Validation_Y = Validation.iloc[:,81]
Result_Train = {}
Result_Valid = {}

for l1 in lambda_1:
  for l2 in lambda_2:
    alpha, l1_ratio = cal_lambdas(l1,l2)
    elastic = ElasticNet(alpha, l1_ratio)
    ml_fit = elastic.fit(Train_X,Train_Y)
    
    train_res = ml_fit.predict(Train_X)
    train_MSE = mean_squared_error(train_res, Train_Y)
    Result_Train[(l1,l2)] = train_MSE
    
    valid_res = ml_fit.predict(Validation_X)
    valid_MSE = mean_squared_error(valid_res, Validation_Y)
    Result_Valid[(l1,l2)] = valid_MSE
    
```

```python
result = pd.DataFrame(columns=['lambda_1','lambda_2','MSE_Train','MSE_Validation'])
for i in lambda_1:
  for d in lambda_2:
    alpha, l1_ratio = cal_lambdas(i,d)
    temp = {'lambda_1': i, 'lambda_2': d, 'MSE_Train': Result_Train[i,d], 'MSE_Validation': Result_Valid[i,d]} 
    result = result.append(temp, ignore_index=True)

pd.set_option('display.max_rows', result.shape[0]+1)
#result
result.to_csv(r'C:\Users\Reza\Desktop\result.csv', index = False)
```

```python
Train_min = min(Result_Train, key=Result_Train.get)
Valid_min = min(Result_Valid, key=Result_Valid.get)
Result_Train_min= Result_Train[Train_min]
Result_Valid_min= Result_Valid[Valid_min]
print("Train data:         lambda1= " + str(Train_min[0]) + " ,  lambda2= "  + str(Train_min[1]) + "   with the MSE:  " + str(Result_Train_min))
print("Validation data:    lambda1= " + str(Valid_min[0]) + " ,  lambda2= "  + str(Valid_min[1]) + "   with the MSE:  "  + str(Result_Valid_min))
```
